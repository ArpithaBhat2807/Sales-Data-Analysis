```python
#To provide labels,automatically gives count as y and we have given gender as x
#But we dont see exact count value.
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
#We get the exact count and we see that number of females>no of males
#Infer: No of purchases made by females is more
```

```python
#To get overall purchase amount gender wise
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_genderwise=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_genderwise)
```

```python
sns.countplot(x='Age Group',data=df)
#This will give the count of purchase in each age group but wont tell how much purchase in that group is done by men and women
```

```python
#Now for each age group,we need purchase done by men and women separately.So we use hue
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
#This is to give count of purchases done in each group gender wise
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Get total spend amount for an age group and plot it
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
df.columns
```

```python
#Get statewise sales count.How much sales is done in each state.So we group by state and get 10 top selling states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,15)})
ax=sns.barplot(x='State',y='Orders',data=sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Get statewise sales amount.How much sales is done in each state.So we group by state and get 10 top selling states amount wise
#A state may have produced more no of orders but the amount totally may still be less
sales_state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,15)})
ax=sns.barplot(x='State',y='Amount',data=sales_state_amount)
for bars in ax.containers:
    ax.bar_label(bars)
```
